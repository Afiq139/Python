# 011 Euler's Formula Exercise

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def eulerFromCosSin():
    re = eval(input('cosine part: '))
    im = eval(input('sine part: '))

    m = np.sqrt(re**2 + im**2)
    k = np.arctan2(im,re)

    plt.polar([0,k],[0,m])
    plt.title('me$^{i\\phi}$, m=%g,$\\phi$=%g' %(m,k))
    plt.thetagrids([0,45,130,200,222.2])
    plt.show()
```

```python
eulerFromCosSin()
```
