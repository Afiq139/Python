```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1.) Plot a series of random numbers
s = np.random.randn(100,1)
plt.plot(s,'s')
plt.show()
```

```python
# 2.) create an image of a matrix of random integers between (and including) 3 and 20

mat = np.random.randint(3,21,(30,20))

print(np.max(mat))
plt.imshow(mat)
plt.colorbar()
plt.show()
```

```python
# 3.) create 100 random phase angle [0,2pi] and show unit vectors with those angles

n = 100
randphases = np.random.rand(n)*2*np.pi

for i in range(0,n):
    plt.polar([0,randphases[i]],[0,1],'o-', color=np.random.rand(3), markersize=20,alpha=.3)

plt.show()
```

```python
# 4.) create an outwards spiral using phase angles and amplitudes

n = 100
a = np.linspace(0,1,n)
p = np.linspace(0,4*np.pi,n)

plt.polar(p,a,'m', linewidth=3)
plt.axis('off')
plt.show()
```

```python
# 5.) Convert radians to degrees

n = 10
rad = np.logspace(np.log10(.001),np.log10(2*np.pi), n)
print(np.rad2deg(rad))
```

```python
# 6.) famous equality in trigonometry

ang = np.logspace(np.log10(.00000001),np.log10(2*np.pi), 10)
np.cos(ang)**2 + np.sin(ang)**2
```
