# 008 Graphing resolution for sine, cosine, and tangent

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 6*np.pi, 10)

plt.plot(x,np.sin(x),'r',label='$\\sin(\\theta)$')
plt.plot(x,np.cos(x),'b',label='$\\cos(\\theta)$')

plt.legend()
plt.xlabel('rad.')
plt.ylabel('function value')
plt.show()
```

```python
th = np.linspace(0,4*np.pi, 100000)

plt.plot(th,np.tan(th),'k')
plt.ylim([-5,5])
plt.show()
```
