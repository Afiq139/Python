# 009 Graphing and resolution exercise

```python
import numpy as np
import matplotlib.pyplot as plt
```

Exercise 1

```python
x = np.linspace(-2*np.pi,2*np.pi,200)

y1 = np.sin(x + np.cos(x)**2)
y2 = np.cos(x + np.sin(x))

plt.plot(x,y1,x,y2)
plt.legend(['sin(x+cos(x)**2)','cos(x+sin(x))'])
plt.show()
```
