# 002 Introduction to random numbers

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nums = np.random.rand(1000)

# nums = nums + 3  # starting from 3 at y-axis
minval = 2
maxval = 17

# nums = nums*(maxval-minval)  # range to 15
nums = nums*(maxval-minval) + minval # range to 15 add 2 = 17
plt.plot(nums, 's')
plt.show()

plt.hist(nums)
plt.show()
```

```python
nums = np.random.randn(1000)  #normal distribution

plt.plot(nums, 's', alpha =.5) #alpha - transparency
plt.show()

plt.hist(nums)
plt.show()
```
