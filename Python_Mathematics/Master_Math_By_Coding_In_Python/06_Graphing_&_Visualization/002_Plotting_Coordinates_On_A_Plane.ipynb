# 002 Plotting Coordinates On A Plane

```python
import matplotlib.pyplot as plt
```

```python
x = 3
y = 5

plt.plot(x,y,'ro')  # r = red, o = circle (also can do g = green, s = square)
# (k = black , p = pentagon)
plt.axis('square') # make the plot in square size
plt.axis([-6,6,-6,6])  # x-axis limits (2 numbers), y-axis limits (2 numbers)
plt.grid()  #adding grid in plot
 
plt.show() #display the final version of the figure
```

```python
x = [-4,5,2,4,5,6,3]
y = [0,3,4,5,2,-3,4]

for i in range(0, len(x)):

    plt.plot(x[i],y[i], 'o', label='point %s' %(1+i))

plt.axis('square')
plt.grid()
plt.legend() #turn on legend on plot
plt.show()
```

```python
plt.plot(4,2,'r+')

axis = plt.gca() #gca = get current access (going to output an object that points to the axis)

ylim = axis.get_ylim()  # limit in y-axis
print(ylim)

axis.set_ylim([ylim[0], 5.39485739]) # set the new y-axis

plt.grid()
plt.show()
```
