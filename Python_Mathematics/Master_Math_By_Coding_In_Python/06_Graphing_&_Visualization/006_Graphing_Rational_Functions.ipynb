# 006 Graphing Rational Functions

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = range(-3,4)
y = np.zeros(len(x))

for i in range(0, len(x)):
    y[i] = 2 - x[i]**2

plt.plot(x,y,'s-')
plt.show()
```

```python
x = np.linspace(-3,4,100) #lower bound, upper bound, number of steps between upper and lower bound
y1 = 2 + abs(x)**(1/3)
y2 = 2 + abs(x)**(1/2)

plt.plot(x,y1,'m-')
plt.plot(x,y2,'k-')

plt.show()
```
