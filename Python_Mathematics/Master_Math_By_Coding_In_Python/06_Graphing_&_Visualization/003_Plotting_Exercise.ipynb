```python
import matplotlib.pyplot as plt
import sympy as sym
from IPython.display import display, Math
```

```python
x = sym.symbols('x')
y = x**2 -3*x

for pnti in range(-11,12):
    plt.plot(pnti, y.subs(x,pnti), 'o')

plt.xlabel('x')
plt.ylabel('f(x) = %s' %(sym.latex(y)))
plt.show()
```

can use the sympy.pretty function to format the expression nicely.
By using sym.pretty(y, use_unicode=True), the expression will be displayed in a more visually appealing mathematical format. Make sure to have the use_unicode=True argument to enable the use of Unicode characters for mathematical symbols. This should give a properly formatted ylabel for the plot.

```python
x = sym.symbols('x')
y = x**2 - 3*x

for pnti in range(-11, 12):
    plt.plot(pnti, y.subs(x, pnti), 'o')

plt.xlabel('x')

# Use sympy.pretty to format the expression nicely
pretty_y = sym.pretty(y, use_unicode=True)
plt.ylabel(f'f(x) = {pretty_y}')

plt.show()
```

 enclose the label text within r'$' to indicate that it's a LaTeX expression, and then we use LaTeX formatting for the mathematical symbols. 

 In Python, when we see a string literal prefixed with r, such as r'$f(x) = x^2 - 3x$', it denotes a "raw string." In a raw string, backslashes are treated as literal characters and are not used to escape special characters. This is especially useful when working with strings that contain backslashes and special characters like LaTeX expressions, regular expressions, or file paths.

```python
x = sym.symbols('x')
y = x**2 - 3*x

for pnti in range(-11, 12):
    plt.plot(pnti, y.subs(x, pnti), 'o')

plt.xlabel('x')

# Manually format the label using LaTeX syntax
plt.ylabel(r'$f(x) = x^2 - 3x$')

plt.show()
```
