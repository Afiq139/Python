# 022 Even and Odd Functions

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5,5,20)

fEven = x**2
fEvenNeg = (-x)**2

plt.plot(x,fEven)
plt.plot(x,fEvenNeg, 'ro')

plt.show()
```

```python
fodd = x**3
foddNeg = (-x)**3
fNegOdd = -fodd

plt.plot(x,fodd)
plt.plot(x,foddNeg, 'ro')
plt.plot(x,fNegOdd, 'g')
plt.show()
```
