# 015 Exercise a family of Gaussians

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
h = .999
t = np.linspace(-3,3,1001)
```

```python
N = 100
G = np.zeros((N,len(t)))
h = np.zeros(N)

for i in range(N):
    h[i] = (i+1)/N
    G[i,:] = np.exp(-4*np.log(2)*t**2/h[i]**2)

plt.pcolormesh(t,h,G)
plt.xlabel('t')
plt.ylabel('h')
plt.show()
```
