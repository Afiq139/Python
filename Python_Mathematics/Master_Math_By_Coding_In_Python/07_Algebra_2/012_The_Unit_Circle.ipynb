# 012 The Unit Circle

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2*np.pi, 100)

plt.plot(np.cos(x), np.sin(x), 'k')
plt.plot([-1.1,1.1], [0,0], '--', color=[.8,.8,.8])
plt.plot([0,0],[-1.1,1.1], '--', color=[.8,.8,.8])

angle= np.pi/4

plt.axis('square')
plt.axis([-1.1, 1.1, -1.1, 1.1])
plt.xlabel('cos(x)')
plt.ylabel('sin(x)')

plt.show()
```
