# 013 Natural Exponent and Logarithm

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Exercise

```python
x = np.linspace(.0001,10,10)

y1 = np.log(np.exp(x))
y2 = np.exp(np.log(x))

plt.plot(x,y1, label='$\\log(e^x)$')
plt.plot(x,y2,'o', label='$e^{\\log(x)}$')

plt.axis('square')
plt.legend()
plt.show()
```
