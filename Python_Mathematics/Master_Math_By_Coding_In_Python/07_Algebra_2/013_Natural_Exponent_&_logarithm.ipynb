# 013 Natural Exponent and Logarithm

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-3,3,99)

plt.plot(x,np.exp(x))
plt.xlabel('x')
plt.ylabel('$e^x$')
plt.show()

plt.plot(x,np.log(x))
plt.xlabel('x')
plt.ylabel('log(x)')
plt.show()
```

# Exercise

```python
x = np.linspace(.0001,10,10)

y1 = np.log(np.exp(x))
y2 = np.exp(np.log(x))

plt.plot(x,y1, label='$\\log(e^x)$')
plt.plot(x,y2,'o', label='$e^{\\log(x)}$')

plt.axis('square')
plt.legend()
plt.show()
```
