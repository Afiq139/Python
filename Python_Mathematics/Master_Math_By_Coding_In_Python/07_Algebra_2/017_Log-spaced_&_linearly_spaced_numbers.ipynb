# 017 Log-spaced and linearly spaced numbers

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.linspace(1,2,10)
```

```python
a = np.log10(1)
b = np.log10(2)

np.logspace(a,b,10)
```

# Exercise

```python
a = 2 #40
b = 100 #45
n = 50

li = np.linspace(a,b,n)
lo = np.logspace(np.log10(a), np.log10(b), n)

plt.plot(li,li,label='linear')
plt.plot(li,lo,label='log')

plt.legend()
plt.axis('square')
plt.show()
```
