# 002 Asteroid Radial Curve

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(-6*np.pi, 6*np.pi, 1000)

a = t

x = a*np.cos(t)**3
y = a*np.sin(t)**3

plt.plot(x,y,'m', linewidth=3)
plt.axis('off')
plt.axis('square')
plt.show()
```

# Exercise

```python
fig,ax = plt.subplots(3,3,figsize=(10,6))
row, col = np.indices((3,3))

for i in range(9):
    a = t**i

    x = a*np.cos(t)**3
    y = a*np.sin(t)**3

    r = row.ravel()[i]
    c = col.ravel()[i]

    ax[r,c].plot(x,y,'k')
    ax[r,c].axis('off')
    ax[r,c].axis('square')

plt.show()
```
