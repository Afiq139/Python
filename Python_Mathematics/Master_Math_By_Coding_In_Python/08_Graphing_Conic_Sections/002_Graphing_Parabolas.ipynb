# 002 Graphing Parabolas

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#parameters
a = 1 # 1/4p
h = -10
k = -2
n = 100

# x-axis points to evaluate the function
x = np.linspace(h-5,h+5,n)

# create the function
y = a*(x-h)**2 + k

plt.plot(x,y)

plt.grid()
plt.axis('square')
plt.show()
```

```python
plt.plot(y,x)

plt.grid()
plt.axis('square')
plt.show()
```
