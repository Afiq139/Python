# 005 Graphing Ellipses

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
h = 1
k = 2
a = 2
b = 3

axlim = np.max((a,b)) + np.max((abs(h), abs(k)))
x = np.linspace(-axlim,axlim,100)
X,Y = np.meshgrid(x,x)

# create the function
ellipse = (X-h)**2/a**2 + (Y-k)**2/b**2 - 1

plt.contour(X,Y,ellipse,0)
plt.plot(h,k,'go')
plt.gca().set_aspect('equal')
```
