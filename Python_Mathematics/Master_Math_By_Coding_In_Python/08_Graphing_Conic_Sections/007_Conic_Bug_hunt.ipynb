```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1.) Make a gaussian
x = np.linspace(-2,2,100)

#create the gaussian
X,Y = np.meshgrid(x,x)
gauss2d = np.exp(-(X**2+Y**2))

# and plot
plt.imshow(gauss2d)
plt.axis('off')
plt.show()
```

```python
# 2.) Draw a circle using meshgrid

r = 3

# grid space
x = np.linspace(-r,r,100)
X,Y = np.meshgrid(x,x)

# create the function
Fxy = X**2 + Y**2 - r**2

#draw it
plt.contour(Fxy,0)
plt.gca().set_aspect('equal')

plt.axis('off')
plt.show()
```
