# 013 Visualizing matrix-vector multiplication

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
M = np.array([ [.5,1],
               [1,.5] ])

v = np.array([1,1])
# v = np.random.randn(2)
# v = np.array([1,1])
# v = np.array([1,-1])

Mv = M@v

plt.plot([0,v[0]],[0,v[1]],'b',label='v',linewidth=2)
plt.plot([0,Mv[0]],[0,Mv[1]],'r--',label='Mv',linewidth=2)

plt.axis('square')
plt.axis([-5,5,-5,5])
plt.legend()
plt.grid()

plt.show()
```
