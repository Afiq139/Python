# 015 Eigendecomposition Exercise

```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
# part 1: 2x2 symmetric matrix. plot Mv and lv 

M = np.array([ [1,2],[2,1] ])
d,V = np.linalg.eig(M)

lv = d[0]*V[:,1] # change to 1 without changing d
Mv = M@V[:,1]

plt.plot([0,lv[0]],[0,lv[1]],'b',label='$\\lambda v$',linewidth=4)
plt.plot([0,Mv[0]],[0,Mv[1]],'r:',label='Mv',linewidth=3)

plt.axis('square')
plt.axis([-3,3,-3,3])
plt.legend()
plt.grid()

plt.show()
```
