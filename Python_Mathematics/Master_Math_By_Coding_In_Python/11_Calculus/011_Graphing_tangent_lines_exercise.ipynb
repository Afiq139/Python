# 011 Graphing Tangent Lines Exercise

```python
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt
```

```python
def computetangent(f,xa,bnds):
    #defining the derivative and values at point xa
    df = sym.diff(f)
    fa = f.subs(x,xa)
    dfa = df.subs(x,xa)

    #evaluate the tangent line
    xx = np.linspace(bnds[0], bnds[1], 200)
    return dfa * (xx-xa) + fa
```

```python
x = sym.symbols('x')
f = x**2
xx = np.linspace(-2,2,200)
ffun = sym.lambdify(x,f)(xx)

for a in xx:
    y = computetangent(f,a,xx[[0,-1]])
    plt.plot(xx,y,color=[abs(a)/3, abs(a)/4, abs(a)/2])

plt.plot(xx,ffun)
plt.axis('square')
plt.axis([-2,2,-1,3])
plt.axis('off')
plt.show()
```
