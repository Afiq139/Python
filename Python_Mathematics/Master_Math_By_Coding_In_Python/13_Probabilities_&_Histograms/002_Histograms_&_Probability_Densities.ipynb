# 002 Histograms and Probability Densities

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = [1,2,2,2,2,2,3,4,5,6,6,6,7,8,9,8,6,3]

plt.hist(data)
plt.xlabel('Value')
plt.ylabel('Number of appearances (count)')
plt.show()

histdata = np.histogram(data,bins=np.arange(1,11))
print(histdata)

plt.bar(histdata[1][:-1], histdata[0])
plt.xlabel('value')
plt.ylabel('Number of appearances (count)')
plt.show()
```
