# 003 Euclid's Pythagorean triplets

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
numberrange = np.arange(2,51)

a = []
b = []
c = []

for m in numberrange:
    for n in numberrange:
        a.append( m**2 - n**2)
        b.append(2*m*n)
        c.append(m**2 + n**2)
```

```python
fig,ax = plt.subplots(1,figsize=(10,7))

plt.scatter(a,b,c=c, marker='o', alpha=.5,linewidth=.5,edgecolors='k')
plt.axis('off')
plt.show()
```

```python
pythcheck = np.array(a)**2 + np.array(b)**2 - np.array(c)**2

plt.plot(pythcheck,'o')
plt.show()
```
